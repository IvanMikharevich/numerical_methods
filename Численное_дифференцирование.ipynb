```
from math import cos, sin, log 
from scipy.misc import derivative
from matplotlib import pyplot as plt
import pylab

x = 3

def f(x):
    return cos(x)*x**(0.5)

def der(x):
  return(-sin(x)*x**(0.5) + cos(x)/(2*x**(0.5)))

def der2(x):
  return(-cos(x)*x**0.5 - cos(x)/(4*x**(3/2)) - sin(x)/(x**0.5))
```

 Вычисление правой разностной производной при различных h.

```
dr_list,  h_list = [], []
for i in range(20):
  h = 10**(-i)
  h_list.append(h)
  dr_list.append(abs(der(x) - (f(x + h) - f(x)) / h))

h_list = [log(x) for x in h_list]
dr_list = [log(x) for x in dr_list]

plt.plot(h_list, dr_list)
```

Вычисление центральной разностной производной при различных h.

```
dr_list,  h_list = [], []
for i in range(20):
  h = 10**(-i)
  h_list.append(h)
  dr_list.append(abs(der(x) - (f(x + h) - f(x - h)) / (2*h)))

h_list = [log(x) for x in h_list]
dr_list = [log(x) for x in dr_list]

plt.plot(h_list, dr_list)
```

Вычисление 2-ой разностной производной при различных h.

```
dr_list,  h_list = [], []
for i in range(20):
  h = 10**(-i)
  h_list.append(h)
  dr_list.append(abs(der2(x) - (f(x + h) - 2*f(x) + f(x - h)) / (h**2)))

h_list = [log(x) for x in h_list]
dr_list = [log(x) for x in dr_list]

plt.plot(h_list, dr_list)
```
