```python
import sympy
 
x = sympy.Symbol('x')
f_x = sympy.sin(x) / x**2
 
def sympson(left, right, n, function):
    h = (right - left) / (2 * n)
 
    tmp_sum = float(function.subs({x: left})) +\
        float(function.subs({x: right}))
 
    for step in range(1, 2 * n):
        if step % 2 != 0:
            tmp_sum += 4 * float(function.subs({x: left + step * h}))
        else:
            tmp_sum += 2 * float(function.subs({x: left + step * h}))
 
    return tmp_sum * h / 3

def trap(left, right, n, function):
    h = (right - left) / (2 * n)
    s=0.5*(float(function.subs({x: left})) +\
        float(function.subs({x: right})))
    arg=left+h
    while (arg<=right-h):
        s+=float(function.subs({x: arg}))
        arg+=h
    return h*s

smp = sympson(1, 2, 20, f_x)
trp = trap(1, 2, 20, f_x)

print(trp, smp, abs(trp - smp))


_x, _y = [], []
for i in range(200):
  n = i + 50
  _x.append(n)
  _y.append(abs(trap(1, 2, n, f_x) - sympson(1, 2, n, f_x)))

import matplotlib.pyplot as plt
plt.plot(_x, _y)
```
